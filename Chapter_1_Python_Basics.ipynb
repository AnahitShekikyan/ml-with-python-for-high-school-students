**Control flow** is how your program makes decisions. Think of it like a choose-your-own-adventure story — depending on certain conditions, different things happen.

```
Is score >= 70?

   ----+----
   │       │
  YES      NO
   │       │

```

Use `if`, `elif` (else if), and `else` to handle different cases.



````python
score = 90

if score >= 90:
    print("A — Excellent!")
elif score >= 80:
    print("B — Great job!")
elif score >= 70:
    print("C — You passed!")
else:
    print("Keep practicing — you've got this!")
````

### 1.8 Loops in Python 🔁

Loops let you repeat a block of code without writing it over and over again. Imagine having to print every song in a 100-song playlist manually — a loop does it in 2 lines!

#### For Loop
Use a `for` loop when you know **how many times** you want to repeat something.

````python
# print every song in a playlist
playlist = ["Espresso", "Flowers", "Cruel Summer", "Vampire", "Karma"]

for song in playlist:
    print("🎵 Now playing:", song)
````

````python
# use range() to repeat something a set number of times
for i in range(5):
    print("Level", i + 1, "complete! ⭐")
````

#### While Loop
Use a `while` loop when you want to keep going **until a condition changes**.

````python
# a game countdown timer
countdown = 5

while countdown > 0:
    print("Game starts in:", countdown)
    countdown -= 1  # same as countdown = countdown - 1

print("🎮 GO!")
````

#### Break and Continue

- **`break`** → exit the loop immediately (like pressing Stop)
- **`continue`** → skip this round and go to the next (like pressing Skip)

````python
# break: stop searching once we find what we want
songs = ["Espresso", "Flowers", "Cruel Summer", "Vampire", "Karma"]

for song in songs:
    if song == "Cruel Summer":
        print("Found it! 🎯 Playing:", song)
        break
    print("Skipping:", song)
````

````python
# continue: skip songs marked as explicit
songs = ["Espresso", "(explicit)", "Cruel Summer", "(explicit)", "Karma"]

for song in songs:
    if song == "(explicit)":
        print("Skipping explicit track...")
        continue
    print("🎵 Playing:", song)
````

### 1.10 Lambda (Anonymous) Functions ⚡

A **lambda function** is a quick, one-line function — perfect for simple operations you don't need to name or reuse.

```
Regular function:          Lambda version:
def double(x):             double = lambda x: x * 2
    return x * 2
```

Lambda functions are commonly used with `filter()` and `map()` to process lists quickly.

````python
# lambda to calculate a letter grade
get_grade = lambda score: "A" if score >= 90 else ("B" if score >= 80 else ("C" if score >= 70 else "F"))

print(get_grade(95))  # A
print(get_grade(83))  # B
print(get_grade(61))  # F
````

````python
# filter() — keep only the items that match a condition
scores = [45, 78, 92, 60, 85, 55, 99, 70]
passing_scores = filter(lambda s: s >= 70, scores)
print("Passing scores:", list(passing_scores))

# map() — transform every item in a list
prices = [10.00, 25.50, 8.99, 45.00]
discounted = map(lambda p: round(p * 0.80, 2), prices)  # 20% off
print("After 20% discount:", list(discounted))
````

### 1.11 Functions for Lists 📋

Python has powerful built-in tools to work with lists — sort them, filter them, transform them, and combine them.

````python
# len() — how many items are in the list?
top_songs = ["Espresso", "Flowers", "Cruel Summer", "Vampire", "Karma"]
print("Songs in playlist:", len(top_songs))

# sort() — sort in place (modifies original list)
scores = [88, 72, 95, 61, 83]
scores.sort()
print("Sorted scores:", scores)

# sorted() — returns a new sorted list (original unchanged)
scores2 = [88, 72, 95, 61, 83]
print("Sorted (new list):", sorted(scores2))
print("Original unchanged:", scores2)
````

````python
# filter() — keep only items matching a condition
followers = [120, 450, 30, 8900, 200, 15]
popular = filter(lambda x: x >= 200, followers)
print("Popular accounts (200+ followers):", list(popular))

# map() — apply a function to every item
prices = [9.99, 14.99, 4.99, 29.99]
with_tax = map(lambda p: round(p * 1.08, 2), prices)  # 8% tax
print("Prices with tax:", list(with_tax))
````

````python
# reduce() — combine all items into one value
from functools import reduce

weekly_steps = [8200, 10500, 7800, 11000, 9300, 6500, 12000]
total_steps = reduce(lambda x, y: x + y, weekly_steps)
print("Total steps this week:", total_steps)
print("Daily average:", total_steps // 7)
````

### 1.12 Functions for Dictionaries 🗃️

Dictionaries store data as **key-value pairs** — like a real dictionary where you look up a word (key) to find its definition (value). Python has many useful tools for working with them.

````python
# a social media profile stored as a dictionary
profile = {
    "username": "coder_girl_22",
    "followers": 1280,
    "following": 340,
    "posts": 47
}

print("Number of fields:", len(profile))
print("Keys:",   list(profile.keys()))
print("Values:", list(profile.values()))
print("All data:", list(profile.items()))
````

````python
# get() — safely retrieve a value (won't crash if key doesn't exist)
print(profile.get("username"))          # output: coder_girl_22
print(profile.get("bio", "No bio yet")) # output: No bio yet

# update() — add or change key-value pairs
profile.update({"followers": 1500, "bio": "Aspiring ML engineer 🤖"})
print(profile)
````

````python
# pop() — remove a key-value pair and return its value
removed = profile.pop("following")
print("Removed value:", removed)
print("Updated profile:", profile)

# in — check if a key exists
print("username" in profile)   # True
print("location" in profile)   # False
````

### 1.13 String Manipulation 🔤

Strings are how Python handles text. Nearly every app — social media, search engines, chatbots — works with strings constantly. Python has tons of built-in tools to slice, search, and transform them.

````python
username = "coder_girl_22"

print("Length:",     len(username))           # number of characters
print("Uppercase:",  username.upper())         # all caps
print("Capitalize:", username.capitalize())    # first letter capital
print("Has digits:", username.isalpha())       # False — has numbers and _
````

````python
# concatenation and repetition
first = "Taylor"
last = "Swift"
full = first + " " + last
print(full)
print("⭐" * 10)   # repeat a string

# slicing — extract part of a string
song = "Cruel Summer"
print(song[0:5])   # first 5 characters
print(song[6:])    # from index 6 to end
print(song[-6:])   # last 6 characters
````

````python
# replace() — swap one word for another
tweet = "I love Mondays!"
fixed_tweet = tweet.replace("Mondays", "Fridays")
print(fixed_tweet)

# split() — break a string into a list
csv_data = "Alex,92,Basketball,11th Grade"
fields = csv_data.split(",")
print(fields)

# format() — insert variables into a string
name = "Alex"
score = 92
print("Student {} scored {} on the test.".format(name, score))

# join() — combine a list back into a string
hashtags = ["Python", "MachineLearning", "HighSchoolCoder"]
print(" #".join(hashtags))
````

### 1.14 Exception Handling 🛡️

Errors happen — even to professional developers! **Exception handling** lets your program deal with errors gracefully instead of crashing.

Think of it like a safety net:
- `try` → attempt the risky code
- `except` → catch the error if it happens
- `finally` → always runs, no matter what (optional)


````python
# without exception handling — this would crash!
# print(10 / 0)  ← ZeroDivisionError

# with exception handling — safe!
try:
    result = 10 / 0
    print(result)
except ZeroDivisionError:
    print("❌ Error: Can't divide by zero!")
````

````python
# handling multiple types of errors
try:
    age = int("sixteen")   # can't convert word to int!
    print(age)
except ValueError:
    print("❌ ValueError: Please enter a number, not a word.")
except TypeError:
    print("❌ TypeError: Wrong type of data.")
finally:
    print("✅ This always runs — error or not.")
````

### 1.15 File Handling in Python 📁

Python can read from and write to files on your computer — like a diary your program can write in and read back later.

| Mode | Meaning |
|------|---------|
| `'r'` | Read (file must already exist) |
| `'w'` | Write (creates new file or overwrites existing) |
| `'a'` | Append (adds to end of existing file) |

````python
# write to a file (creates it if it doesn't exist)
file = open('my_journal.txt', 'w')
file.write("Day 1: Started learning Python today!")
file.close()
````

````python
# append — add a new entry without erasing the old one
file = open('my_journal.txt', 'a')
file.write("\nDay 2: Learned about loops and functions!")
file.close()
````

````python
# read with 'with' statement — best practice, auto-closes the file
with open('my_journal.txt', 'r') as file:
    contents = file.read()
    print(contents)
````

````python
# readline() — read one line at a time
with open('my_journal.txt', 'r') as file:
    first_line = file.readline()
    print("First entry:", first_line)
````
